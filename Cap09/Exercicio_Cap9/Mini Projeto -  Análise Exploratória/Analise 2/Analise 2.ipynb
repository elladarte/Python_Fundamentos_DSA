```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "C:/PythonFundamentos/Exercicios Feitos/Data Science Academy/Cap9/Mini Projeto -  Análise Exploratória/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
df.columns
```

# Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
```

```python
sns.set_style("whitegrid")
g = sns.factorplot(y="brand", data=df, kind="count", size=7, aspect=1.5)
g.ax.set_title("Veículos Por Marca",fontdict={'size':18})
g.ax.xaxis.set_label_text("Número de Veículos",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Marca",fontdict= {'size':16})
plt.show()
```

```python
g.savefig("brand-vehicleCount.png")
```

# Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
```

```python
fig2, ax = plt.subplots(figsize=(8,5))
sns.barplot(x="vehicleType", y="price", hue="gearbox", data=df)
ax.set_title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio",fontdict={'size':18})
ax.xaxis.set_label_text("Tipo de Veículos",fontdict= {'size':16})
ax.yaxis.set_label_text("Preço Medio",fontdict= {'size':16})
plt.show()
```

```python
fig2.savefig("vehicletype-gearbox-price.png")
```
