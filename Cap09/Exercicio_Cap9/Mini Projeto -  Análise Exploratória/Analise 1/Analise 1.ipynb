# Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

# Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
```

```python
df.columns
```

```python
fig,ax = plt.subplots(figsize=(8,6))
sns.distplot(df["yearOfRegistration"], color="#33cc33",kde=True, ax=ax)
plt.title("Distribuição de Veículos com base no Ano de Registro",fontsize=15)
plt.ylabel("Densidade (KDE)",fontsize=15)
plt.xlabel("Ano de Registro",fontsize=15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("vehicle-distribution.png")
```

# Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
```

```python
sns.set_style("whitegrid")
fig2, ax2 = plt.subplots(figsize=(8,6))
sns.boxplot(x="vehicleType", y="price", data=df)
ax2.set_title("Análise de Outliers",fontsize=15)
ax2.xaxis.set_label_text("Tipo de Veículo", fontsize=15)
ax2.yaxis.set_label_text("Range de preço", fontsize=15)
plt.show()
```

```python
fig2.savefig("price-vehicleType-boxplot.png")
```

# Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
```

```python
fig3, ax3 = plt.subplots(figsize=(8,6))
sns.countplot(x="vehicleType",data=df)
ax3.set_title("Contagem total de veículos à venda conforme o tipo de veículo",fontsize=15)
ax3.xaxis.set_label_text("Tipo de Veículo", fontsize=15)
ax3.yaxis.set_label_text("Total de Veículos para vendas", fontsize=15)
plt.show()
```

```python
fig3.savefig("count-vehicleType.png")
```
