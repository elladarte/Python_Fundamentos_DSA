```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "C:/PythonFundamentos/Exercicios Feitos/Data Science Academy/Cap9/Mini Projeto -  Análise Exploratória/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

# Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x="fuelType", y="price", hue="gearbox", data=df)
ax.set_title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio",fontdict={'size':18})
ax.xaxis.set_label_text("Tipo de Combustivel",fontdict= {'size':16})
ax.yaxis.set_label_text("Preço Medio",fontdict= {'size':16})
plt.show()
```

```python
fig.savefig("fueltype-vehicleType-price.png")
```
